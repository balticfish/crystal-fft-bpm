```python
full_run1 = np.load(datafile1)
# full_run2 = np.load(datafile2)
# full_run3 = np.load(datafile3)


n_cycle, n_par, n_data = np.shape(full_run1)
```

```python
XOPfile = 'Si4009keVThickness50um'
XOPdata = np.loadtxt(XOPfile)
```

```python
min_theta = np.min(XOPdata[:,0])
max_theta = np.max(XOPdata[:,0])
```
