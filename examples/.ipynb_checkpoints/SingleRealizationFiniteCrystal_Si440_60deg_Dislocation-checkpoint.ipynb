```python

def single_realization_with_data(delta_theta,Rock_angle):
    
    xcr = XCrystal(XCr_path+'/config/Si440_17p45keVDislk60degGronkowskiFiniteCrystal3D.yaml',omega)
    xcr.configure(delta_theta,Rock_angle)
    xcr.run3D()
    return xcr.Reflectivity, xcr.Transmission, xcr.U1_field, xcr.U2_field, xcr.u, xcr.M
```

```python
omega0=17450
```

```python
U1_data = []
U2_data = []

Rs, Ts, U1_xyz, U2_xyz, up, MM = single_realization_with_data(8.0e-6,omega0)
U1_data.append(U1_xyz)
U2_data.append(U2_xyz)

U1_xyz = np.asarray(U1_data)
U2_xyz = np.asarray(U2_data)
```

```python
def my_pad( wavefront, shape):
        """
        Pad an array with complex zero elements.

        Parameters
        ----------
        wavefront: np.ndarray
            The array to pad
        shape: array_like
            Number of values padded to the edges of each axis

        Returns
        -------
        np.ndarray

        """

        return np.pad(wavefront, shape, mode='constant', constant_values=(0.0 + 1j*0.0, 0.0 + 1j*0.0))
```

```python
a= U1_xyz[0,:,:,xcr.M-1]
a.shape
padshape=[600,400]
a=my_pad(a,padshape)
a.shape[0]
dxx=xcr.xx[1]-xcr.xx[0]
dyy=xcr.yy[1]-xcr.yy[0]
xxp = dxx*(np.arange(1,a.shape[0]+1) - 0.5*(a.shape[0]))
yyp = dxx*(np.arange(1,a.shape[1]+1) - 0.5*(a.shape[1]))
xxp.shape
xcr.xx.shape
yyp.shape
Yyp, Xxp = np.meshgrid(yyp,xxp) # x,y mesh/grid
Yyp.shape
```

```python
L=2*xcr.convr
r_sq=Xxp**2+Yyp**2
G11 = np.fft.fftshift(np.fft.fft2(a*np.exp(1j/2*r_sq/L)), axes=(0,1))
G11.shape
plt.imshow(np.log(np.abs(G11+20)**2))
plt.imshow((np.abs(G11+20)**0.2))
plt.ylim(400,800)
plt.xlim(400,800)
```
