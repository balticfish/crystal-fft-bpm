```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import time
XCr_path = os.getcwd()+'/../'
sys.path.append(XCr_path)
from XCrystal import *
t0 = time.time()
```

```python
omega= 17450.0
xcr = XCrystal(XCr_path+'/config/Si440_17p45keVDislk60degGronkowskiFiniteCrystal3D.yaml',omega)
pp=xcr.configure(0,0)
```

```python
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
U1fileN = 'run_omega_U1_1_36_angleFiniteCrystal_Si440_60deg_No_Dislocation.npy'
U2fileN = 'run_omega_U2_1_36_angleFiniteCrystal_Si440_60deg_No_Dislocation.npy'

U1_wxyzN = np.load(U1fileN)
U2_wxyzN = np.load(U2fileN)
```

```python
U1file = 'run_omega_U1_1_36_angleFiniteCrystal_Si440_60deg_Dislocation.npy'
U2file = 'run_omega_U2_1_36_angleFiniteCrystal_Si440_60deg_Dislocation.npy'

U1_wxyz = np.load(U1file)
U2_wxyz = np.load(U2file)
```

```python

aN=U1_wxyzN[:,:,:,5]
aN=np.pad(aN, ((0, 0),(500,500), (500,500)), 'constant', constant_values=(0.0 + 1j*0.0, 0.0 + 1j*0.0))
 
```

```python

a=U1_wxyz[:,:,:,5]
a=np.pad(a, ((0, 0),(500,500), (500,500)), 'constant', constant_values=(0.0 + 1j*0.0, 0.0 + 1j*0.0))
 
```

```python
a.shape
```

```python
dxx=xcr.xx[1]-xcr.xx[0]
dyy=xcr.yy[1]-xcr.yy[0]
xxp = dxx*(np.arange(1,a.shape[1]+1) - 0.5*(a.shape[1]))
yyp = dxx*(np.arange(1,a.shape[2]+1) - 0.5*(a.shape[2]))
xxp.shape
xcr.xx.shape
yyp.shape
Yyp, Xxp = np.meshgrid(yyp,xxp) # x,y mesh/grid
Yyp.shape
```

```python
L=2*xcr.convr
r_sq=Xxp**2+Yyp**2
G11N = np.fft.fftshift(np.fft.fft2(aN*np.exp(1j/2*r_sq/L)),axes=(1,2))
G11 = np.fft.fftshift(np.fft.fft2(a*np.exp(1j/2*r_sq/L)),axes=(1,2))
```

```python
fig, ax = plt.subplots(nrows=6, ncols=6)
ii=-1
for row in ax:
    for col in row:
        ii=ii+1
        col.imshow((np.abs(G11[ii,:,:])**0.5),cmap='jet')

for ax in fig.get_axes():
    ax.label_outer()
    ax.set_ylim(550,750)
    ax.set_xlim(450,700)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=6, ncols=6)
ii=-1
for row in ax:
    for col in row:
        ii=ii+1
        col.imshow((np.abs(G11N[ii,:,:])**0.5),cmap='jet')

for ax in fig.get_axes():
    ax.label_outer()
    ax.set_ylim(550,750)
    ax.set_xlim(450,700)
plt.show()
```
