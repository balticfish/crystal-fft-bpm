```python
import sys
import os
XCr_path = os.getcwd()+'/../'
sys.path.append(XCr_path)
from XCrystal import *
import XCrTools as tools
omega0 = 12800.0

xcr = XCrystal(XCr_path+'/config/C400_TD_9p8keV_ThermalBump1.yaml', omega0)
xtools = tools.XCrTools(xcr)

```

```python
U1file = 'run_omega_U1_46_16_C400_9831eVBumpCopy1-x0s55.npy'
U2file = 'run_omega_U2_46_16_C400_9831eVBumpCopy1-x0s55.npy'

U1_wxyz = np.load(U1file)
U2_wxyz = np.load(U2file)
```

```python
U1_wxyz.shape
```

```python
IJ=150
```

```python
#extent = [0, xcr.xxmax / xcr.convr * 1e6, np.max(w_sim), np.min(w_sim)]
#GnBu
plt.figure(figsize=(10,8))
plt.imshow(np.abs(U1_wxyz[:, :,0,IJ])**2 + np.abs(U2_wxyz[:,:,0,IJ])**2, aspect='auto')
plt.xlabel(r'X ($\mu$m)')
plt.ylabel(r'$\Delta \omega$ (eV)')

#plt.ylim(0, 730)
```

```python
U1_wx = U1_wxyz[:, :, 0, :]
U2_wx = U2_wxyz[:, :, 0, :]

U1_tx = np.fft.ifft(U1_wx, axis=0)
U2_tx = np.fft.ifft(U2_wx, axis=0)
```

```python
It=90
U1_txIt=U1_tx[:,:,It];
U2_txIt=U2_tx[:,:,It];
plt.figure(figsize=(6,5))
plt.imshow(np.abs(U1_txIt*1+U2_txIt*1)**0.5, aspect='auto', cmap='jet')
plt.ylabel(r'$\Delta t$ (a.u.)')
plt.ylim(400,340)
#plt.grid()
plt.show()
```
