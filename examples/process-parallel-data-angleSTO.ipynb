```python
import sys
import os
XCr_path = os.getcwd()+'/../'
sys.path.append(XCr_path)

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from XCrystal import *
import time
```

```python
matplotlib.rcParams['figure.figsize'] = (12,8)
matplotlib.rcParams['axes.labelsize'] = 20
matplotlib.rcParams['xtick.labelsize'] = 20
matplotlib.rcParams['ytick.labelsize'] = 20
matplotlib.rcParams['legend.fontsize'] = 20
```

```python
datafile1 = XCr_path +'/examples/run_16_16_angleSTO_10keVZ0p002alpha_0em3wsubstrate.npy'
datafile3 = XCr_path +'/examples/SpectrumReflSTO71124Strain.dat'
datafile2 = XCr_path +'/examples/SpectrumAnglSTO71124Strain.dat'
datafile5 = XCr_path +'/examples/SpectrumReflSTO71124StrainWsubstrate.dat'
datafile4 = XCr_path +'/examples/SpectrumAnglSTO71124StrainWsubstrate.dat'
datafile6 = XCr_path +'/examples/run_8_16_angleSTO_10keVZ0p002alpha_0em3wsubstratexmax0p5um.npy'
```

```python
full_run1 = np.load(datafile1)
n_cycle, n_par, n_data = np.shape(full_run1)
Matangle = np.loadtxt(datafile2)
MataRefl = np.loadtxt(datafile3)
Matangle2 = np.loadtxt(datafile4)
MataRefl2 = np.loadtxt(datafile5)
full_run6 = np.load(datafile6)

n_cycle6, n_par6, n_data6 = np.shape(full_run6)
```

```python
RCI_data1 = np.reshape(full_run1, (n_cycle * n_par, n_data))
RCI_data6 = np.reshape(full_run6, (n_cycle6 * n_par6, n_data6))
```

```python
plt.plot(1e6 * RCI_data1[:,0], RCI_data1[:,1], '-', linewidth=2.1, color='g', label='python')
plt.plot(1e6 *Matangle[:] ,MataRefl[:], '-', linewidth=2.1, color='r', label='Matlab')
plt.plot(1e6 *Matangle2[:] ,MataRefl2[:], '+-', linewidth=2.1, color='b', label='Matlab')
#plt.plot(XOPdata[:,0]*4.84813, XOPdata[:,1], '-', linewidth=2.1, color='red', label='XOP')
plt.plot(1e6 * RCI_data6[:,0], RCI_data6[:,1], 'o-', linewidth=2.1, color='b', label='pythonxxmx2um')
plt.legend()
plt.xlabel(r'$\Delta \theta$ (urad)')
plt.ylabel('R')
#plt.ylim(-0.1, 1.5)
#plt.xlim(-6000, -5000)
plt.yscale("log")
```
