##  Simulations of strongly bent diamond crystal used as a dispersive spectrometer for XFEL pulses
### Angular scan

```python
matplotlib.rcParams['figure.figsize'] = (8,4)
matplotlib.rcParams['axes.labelsize'] = 20
matplotlib.rcParams['xtick.labelsize'] = 20
matplotlib.rcParams['ytick.labelsize'] = 20
matplotlib.rcParams['legend.fontsize'] = 20
```

## Loading and reshaping angular scan data

```python
datafile1 = XCr_path +'/examples/run_2_64_angleC440_14p4keVR65.npy'
datafile2 = XCr_path +'/examples/run_2_64_angleC440_14p4keVR95.npy'
```

```python
full_run1 = np.load(datafile1)
n_cycle, n_par, n_data = np.shape(full_run1)
full_run2 = np.load(datafile2)
n_cycle2, n_par2, n_data2 = np.shape(full_run1)
```

```python
RCI_data1 = np.reshape(full_run1, (n_cycle * n_par, n_data))
RCI_data2 = np.reshape(full_run2, (n_cycle2 * n_par2, n_data2))
```

## Displaying results

```python
plt.plot(1e6 * RCI_data1[:,0], RCI_data1[:,1]/np.max(RCI_data1[:,1])+0.4, '.-', linewidth=2.5, color='g', label='R=65 mm')
plt.plot(1e6 * RCI_data2[:,0], RCI_data2[:,1]/np.max(RCI_data2[:,1]), '.-', linewidth=2.5, color='r', label='R=95 mm')
#plt.plot(XOPdata[:,0]*4.84813, XOPdata[:,1], '-', linewidth=2.1, color='red', label='XOP')
plt.legend()
plt.xlabel(r'$\Delta \theta$ (urad)')
plt.ylabel('R')
plt.ylim(-0.1, 1.5)
#plt.xlim(-6, 6)
```
