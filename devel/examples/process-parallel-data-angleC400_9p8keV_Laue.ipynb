### Calculating rocking curve for C(400) reflection in Laue geometery at 9.831 keV for 82 $\mu$m thick crystal

```python
matplotlib.rcParams['figure.figsize'] = (10,7)
matplotlib.rcParams['axes.labelsize'] = 20
matplotlib.rcParams['xtick.labelsize'] = 20
matplotlib.rcParams['ytick.labelsize'] = 20
matplotlib.rcParams['legend.fontsize'] = 20
```

## Loading and reshaping angular scan data

```python
datafile = XCr_path +'/examples/run_8_16_angleC400_9p8keV_Laue_waist2000um_thick_82_90p00deg.npy'
```

```python
XOPfile =XCr_path + '/examples/C400Laue9p8keV'
XOPdata = np.loadtxt(XOPfile)
```

```python
full_run = np.load(datafile)

n_cycle, n_par, n_data = np.shape(full_run)
n_cycle, n_par, n_data
```

```python
min_theta = np.min(XOPdata[:,0])
max_theta = np.max(XOPdata[:,0])
```

```python
min_theta, max_theta
```

```python
RCI_data = np.reshape(full_run, (n_cycle * n_par, n_data))
```

## Displaying results

```python
plt.plot(1e6 * RCI_data[:,0]+0.0, RCI_data[:,1], '+-', linewidth=4.1, color='b', label='FFT BPM')#plt.plot(1e6 * RCI_data1[:,0]+0.0, RCI_data1[:,1], '+-', linewidth=2.1, color='g', label='FFT BPM')
plt.plot(XOPdata[:,0]*4.84813, XOPdata[:,1], '-', linewidth=4.1, color='red', label='XOP')
plt.legend()
plt.xlabel(r'$\Delta \theta$ (urad)')
plt.ylabel('R')
plt.ylim(0.0, 1)
plt.xlim(-5, 5)
```
